```python
# jointplot ---> Draw a plot of two variables with bivariate and univariate graphs.

from scipy.stats import stats
sns.jointplot(x='fare',y='age',data=titanic,height=7,xlim=(-100,600))
```

```python
# Distribution Plot: The heights of the bars tell us how many data points are in each bin.

sns.displot(titanic['fare'],kde=False,bins=30)
```

```python
#sns.histplot(data=titanic,x='fare',bins=30)
```

```python
titanic.columns
```

```python
# box and whisker plot—also called a box plot—displays the five-number summary of a set of data. 
# The five-number summary is the minimum, first quartile, median, third quartile, and maximum. 
# In a box plot, we draw a box from the first quartile to the third quartile. 
# A vertical line goes through the box at the median.

sns.boxplot(x='class',y='age',data=titanic)
```

```python
# Swarm Plot: Draw a categorical scatterplot with non-overlapping points
# in a swarm plot, the data points don't overlap and are adjusted along the categorical axis. 

plt.figure(figsize=(8,5))
sns.swarmplot(x='class',y='age',data=titanic,size=4)      # default size 5
```

```python
# seaborn.countplot() method is used to Show the counts of observations in each categorical bin using bars.

sns.countplot(x='sex',data=titanic)
```

```python
# A Heat map is a graphical representation of multivariate data that is structured as a matrix of columns and rows.
# Heat maps are very useful in describing correlation among several numerical variables, visualizing patterns and anomalies.
# The heat map transforms the correlation matrix into color coding.
# The correlation matrix shows how the variables are correlated to each other on a scale of -1 to 1, 
# with 1 being a perfect positive correlation and -1 being a perfect inverse correlation.

h = sns.heatmap(titanic.corr(), cmap='coolwarm')
h.set(title='titanic.corr()')
```

```python
# FacetGrid class helps in visualizing distribution of one variable as well as 
# the relationship between multiple variables separately.
# A FacetGrid can be drawn with up to three dimensions ? row, col, and hue
# The variables should be categorical 

graph = sns.FacetGrid(data=titanic,col='sex')
graph.map(sns.histplot,'age',bins=10,color='orange')
```
